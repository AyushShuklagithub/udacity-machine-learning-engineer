# Plot scatter plot and linear regression model

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

plt.style.use('ggplot')

bmi_life_data = pd.read_csv("data/bmi_and_life_expectancy.csv")
bmi_life_data

X_train = bmi_life_data[['BMI']]
y_train = bmi_life_data[['Life expectancy']]

model = LinearRegression()
model.fit(X_train,y_train)

plt.scatter(X_train, y_train)
plt.scatter(X_train, y_train,  color='black')
plt.plot(X_train, model.predict(X_train), color='blue', linewidth=2)

plt.show()
```
