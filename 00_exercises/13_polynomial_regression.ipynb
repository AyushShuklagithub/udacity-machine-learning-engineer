# Load laptop data (charged time & battery life)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

df = pd.read_csv('data/laptop_battery_life.csv', header=None, names=['charged_time', 'battery_life'])
df.head()
```
