# Create terrain data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from common_functions import plot_2d_scatter, make_terrain_data

X, y = make_terrain_data()
plt.style.use('seaborn')
plot_2d_scatter(X, y)
```

# Create AdaBoost model

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

model = AdaBoostClassifier(base_estimator = DecisionTreeClassifier(max_depth=2), n_estimators = 4)
model.fit(X, y)
```

# Plot model with decision boundaries

```python
from common_functions import plot_model_with_boundaries

plot_model_with_boundaries(model, X, y, '')
```
